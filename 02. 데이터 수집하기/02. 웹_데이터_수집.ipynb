### 웹 데이터 수집

```python
import requests, bs4
targetUrl = "https://sparkkorea.com/%ed%80%b4%ec%a6%88/"
resp = requests.get(url = targetUrl)
# if resp.status_code == 200:
### 1-1. 원하는 웹사이트 데이터 수집
htmlObj = resp.text
bsObj = bs4.BeautifulSoup( htmlObj, "html.parser" )
bsObj
### 1-2. 태그정보 수집하기
bsObj.find(name = "div")
sparkQuizDiv = bsObj.find(name="div", attrs={"id":"id_spark_quiz"}     )

textList = []
linkList = []
aLinkTags = sparkQuizDiv.findAll(name="a")
for i in range(0, len(aLinkTags)):
    eachTag = aLinkTags[i]
    # 태그의 텍스트값 불러오기
    eachTagText = eachTag.text
    eachTagLink = eachTag.attrs["href"]
    textList.append( eachTagText )
    linkList.append( eachTagLink )
```

```python
pd.DataFrame( zip(textList, linkList) )
```
